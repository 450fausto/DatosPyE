```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/450fausto/DatosPyE/main/Ligas_Europa/LaLiga_2018_2019.csv')
```

```python
#  GRÁFICO DE PASTEL PARA BASES DE DATOS
columna_interes='FTR' # La variable que me interesa representar en el gráfico de pastel
columna_referencia='Date' # La variable a la cual se le aplica la función de agregación
separacion=0.01 # Separación entre trozos

ax = df.groupby([columna_interes]).count().plot(
                               kind="pie", # Tipo de gráfico: pie chart
                               y=columna_referencia, 
                               ylabel="", # Etiqueta para la variable de referencia
                               xlabel="", # Etiqueta para la variable de interés
                               autopct='%1.0f%%', # %n.mf% n: mínimo de enteros, m: flotantes
                               colors = ['green', 'lightblue', 'orange'], # Colores para el pastel
                               figsize=(10,6), # Tamaño (m,n) m: ancho, n: alto
                               fontsize=12, #Tamaño de letra
                               labels=['' for i in df[columna_interes].unique()], # Eliminar las etiquetas de los trozos del pastel
                               explode=[separacion for i in df[columna_interes].unique()], 
                               )

ax.legend(['Visitante', 'Empate','Local'], # Etiquetas de leyenda
          fontsize=12, # Tamaño de letra en la leyenda
          loc='upper left')  # Localización de la leyenda: 'upper left' 'upper center'
                             # 'upper right' 'lower left' 'lower center' 'lower right'

ax.set_title('Gráfico de pastel', # Título del gráfico
             fontsize=20, # Tamaño de letra del título
             loc='right') # Localización del título: 'left' 'center' 'right'
```

```python
#  GRÁFICO DE PASTEL PARA DATOS CONOCIDOS POR EL USUARIO
indices = ['Carlos', 'Marta', 'Raul', 'Damian', 'Jorge']
valores = [10, 30, 40, 40, 23]
separacion=0.01 # Separación entre trozos

df=pd.DataFrame({'valores':valores}, index=indices)
ax = df['valores'].plot(
            kind="pie", # Tipo de gráfico 
            ylabel="", # Etiqueta para la variable de referencia
            xlabel="", # Etiqueta para la variable de interes
            autopct='%1.0f%%', # %n.mf% n: mínimo de enteros, m: flotantes
            colors = ['green', 'lightblue', 'orange', 'red', 'yellow'], # Colores para el pastel
            figsize=(10,7), # Tamaño (m,n) m: ancho, n: alto
            fontsize=12,
            explode=[separacion for i in df.index], 
            )

ax.legend(loc='upper left', fontsize=12)  # Localización de la leyenda

ax.set_title('Gráfico de pastel', fontsize=20, loc='right') # Título, fuente y posición
df
```
