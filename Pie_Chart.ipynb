```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/450fausto/DatosPyE/main/Ligas_Europa/LaLiga_2018_2019.csv')
```

```python
#  GRAFICO DE PASTEL PARA BASES DE DATOS
columna_interes='FTR' # La variable que me interesa representar en el grafico de pastel
columna_referencia='Date' # La variable a la cual se le aplica la funcion de agregacion
separacion=0.01 # separacion entre trozos

ax = df.groupby([columna_interes]).count().plot(
                               kind="pie", # tipo de grafico
                               y=columna_referencia, 
                               ylabel="", # Etiqueta para la variable de referencia
                               xlabel="", # Etiqueta para la variable de interes
                               autopct='%1.0f%%', # %n.mf% n: minimo de enteros, m: flotantes
                               colors = ['green', 'lightblue', 'orange'], # colores para el pastel
                               figsize=(10,6), # tamano (m,n) m: ancho, n: alto
                               fontsize=12, #tama;o de letra
                               labels=['' for i in df[columna_interes].unique()], # Eliminar las etiquetas de los trozos del pastel
                               explode=[separacion for i in df[columna_interes].unique()], 
                               )

ax.legend(['Visitante', 'Empate','Local'], # Etiquetas de leyenda
          fontsize=12, #tama;o de letra en la leyenda
          loc='upper left')  # localizacion de la leyenda

ax.set_title('GrÃ¡fico de pastel', # titulo del grafico
             fontsize=20, # tama;o de letra del titulo
             loc='right') # localizacion del titulo
```

```python
#  GRAFICO DE PASTEL PARA DATOS CONOCIDOS POR EL USUARIO
indices = ['Carlos', 'Marta', 'Raul', 'Damian', 'Jorge']
valores = [10, 30, 40, 40, 23]
separacion=0.01 # separacion entre trozos

df=pd.DataFrame({'valores':valores}, index=indices)
ax = df['valores'].plot(
            kind="pie", # tipo de grafico 
            ylabel="", # Etiqueta para la variable de referencia
            xlabel="", # Etiqueta para la variable de interes
            autopct='%1.0f%%', # %n.mf% n: minimo de enteros, m: flotantes
            colors = ['green', 'lightblue', 'orange', 'red', 'yellow'], # colores para el pastel
            figsize=(10,7), # tamano (m,n) m: ancho, n: alto
            explode=[separacion for i in df.index], 
            )

ax.legend(loc='upper left')  # localizacion de la leyenda

ax.set_title('GrÃ¡fico de pastel', fontsize=20, loc='right') # Titulo, fuente y posicion
df
```
