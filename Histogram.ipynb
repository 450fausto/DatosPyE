```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/450fausto/DatosPyE/main/heart%20failure/heart.csv")
```

```python
#  GRÁFICO DE BARRAS AGRUPADAS PARA BASES DE DATOS
barras=5     # bins, cantidad de clases
columna_interes='Age'

ax = df[columna_interes].plot(
                kind="hist", # Tipo de gráfico: histogram
                bins=barras, # Cantidad de clases
                figsize=(10,6), # Tamaño (m,n) m: ancho, n: alto
                fontsize=12, #Tamaño de letra
                #ylim=(0,16), # Límite superior e inferior para el eje 'y'
                #yticks=[2,5,8,11,14], # Líneas horizontales
                rot=0, # Inclinación de las etiquetas en grados
                color='gray', # Colección de colores
                )

ax.set_title('Histograma', # Título del gráfico
             fontsize=20, # Tamaño de letra del título
             loc='center') # Localización del título: 'left' 'center' 'right'
ax.set_xlabel('Edades', fontsize=12) # Nombre del eje 'x' y tamaño de letra
ax.set_ylabel('Frecuencia', fontsize=12) # Nombre del eje 'x' y tamaño de letra


ax.grid(axis='y',          # Maya del eje 'y'
        color='gray',      # Color de la maya en eje 'y'
        linestyle='--',    # Estilo de línea
        linewidth=0.5)     # Ancho de línea

ax.grid(axis='x', linewidth=0)  

ancho_de_clase=(df[columna_interes].max()-df[columna_interes].min())/barras
ax.set_xticks(ticks=[df[columna_interes].min()+i*ancho_de_clase for i in range(barras+1)])
```
